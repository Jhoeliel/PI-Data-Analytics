### Ahora usaremos la libreria WORDCLOUD

```python
# importando las librerias necesarias
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generando el wordcloud en memoria
msk = np.array(Image.open('solid-cloud.jpg')) # Podemos usar como mascara cualquier imagen apropiada que deseemos
wc=WordCloud(background_color='black',
             mask=msk,
             collocations=False,
             width=700,
             height=450)
with open('./Moocs/output.txt') as txt_file: # Leemos el texto que vamos a usar
    texto= txt_file.read()

wc.generate(texto) # generamos el wordcloud
```

```python
image_colors = ImageColorGenerator(msk) # usamos los colores de la mascara creada
wc.recolor(color_func=image_colors)

# usaremos matplotlib para generar la imagen
plt.figure(figsize=(20,20))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
wc.to_file('nubeWC.png') # exportamos a archivo .png
plt.show()
```
